```python
import pandas as pd
import json

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('healthcare_dataset.csv')

# Create a list to store examples
examples = []

# Iterate through all rows in the DataFrame
for index, row in df.iterrows():
    # Create an example with input (prompt) and output (completion)
    example = {
        'input': f"Patient {row['Name']} ({row['Age']} years, {row['Gender']}) was admitted to {row['Hospital']} on {row['Date of Admission']} with {row['Medical Condition']}. The doctor in charge was {row['Doctor']}.",
        'output': f"During the admission, the patient was in room {row['Room Number']} and the billing amount was ${row['Billing Amount']:.2f}. The patient was discharged on {row['Discharge Date']} after receiving {row['Medication']} and the test results were {row['Test Results']}."
    }

    # Append the example to the list
    examples.append(example)

# Write the examples to a JSONL file
with open('healthcare_dataset_llama2_all.jsonl', 'w') as jsonl_file:
    for example in examples:
        jsonl_file.write(json.dumps(example) + '\n')
```

```python
with open('healthcare_dataset_llama2_all.jsonl', 'r') as jsonl_file:
    # Read and print the first 5 dictionaries
    for i in range(5):
        line = jsonl_file.readline()
        if not line:
            break  # Break if the end of the file is reached
        # Parse the JSON object
        example = json.loads(line)
        # Print the raw dictionary
        print(f"Example {i + 1}:\n")
        print(example)
        print("=" * 50)
```

```python
import transformers
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Config model
bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Finetuned model
model_finetuned = AutoModelForCausalLM.from_pretrained(
    "GiaGia2002123/healthcare_GPT_llama2",
    device_map="auto",
    load_in_4bit=True,
    quantization_config=bnb_config,
    token = HF_TOKEN ##Put the token here##
)

# Tokenizer
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf",
                                          padding_side="left",
                                          token = HF_TOKEN) ##Put the token here##
tokenizer.pad_token = tokenizer.eos_token  # Most LLMs don't have a pad token by default

# Generation pipeline finetuned
generate_finetuned = pipeline(
    model = model_finetuned,
    tokenizer = tokenizer,
    return_full_text = True,
    task = 'text-generation',
    temperature = 0.1,
    max_new_tokens = 64,
    repetition_penalty = 1.1
)

# ------------------------------------------
# Base model
model_base = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Llama-2-7b-hf",
    device_map="auto",
    load_in_4bit=True,
    quantization_config=bnb_config,
    token = HF_TOKEN
)

# Generation pipeline base
generate_base = pipeline(
    model = model_base,
    tokenizer = tokenizer,
    return_full_text = True,
    task = 'text-generation',
    temperature = 0.1,
    max_new_tokens = 128,
    repetition_penalty = 1.1
)

# Output
print("Finetuned Model:")
input_text = """
Suppose a female at the aged of 50, blood-type: AB, medical conditions is hypertension, what is the test result of this person?
"""
res = generate_finetuned(input_text)
print(res[0]['generated_text'])

print("Base model:")
input_text = """
Suppose a female at the aged of 50, blood-type: AB, medical conditions is hypertension, what is the test result of this person?
"""
res = generate_base(input_text)
print(res[0]['generated_text'])
```
